# Exercise 01 — Basic Operations

```python
import pandas as pd
```

```python
import pandas as pd

views = pd.read_csv(
    'data/feed-views.log',
    sep='\t',
    skiprows=[2, 3],
    skipfooter=2,
    engine='python',
    names=['datetime', 'user']
)

#  datetime ni to'g'ri dtype ga o'tkazish
views['datetime'] = pd.to_datetime(views['datetime'])

```

```python
#  yangi columnlar chiqarish
views['year'] = views['datetime'].dt.year
views['month'] = views['datetime'].dt.month
views['day'] = views['datetime'].dt.day
views['hour'] = views['datetime'].dt.hour
views['minute'] = views['datetime'].dt.minute
views['second'] = views['datetime'].dt.second

views.head()
```

```python
#daytime ni aniqlash

bins = [-1, 3, 6, 10, 16, 19, 23]
labels = [
    'night',
    'early morning',
    'morning',
    'afternoon',
    'early evening',
    'evening'
]

views['daytime'] = pd.cut(views['hour'], bins=bins, labels=labels)
```

```python
#User ustunini indexga aylantirish
views.set_index('user', inplace=True)
```

```python
#elementlari soni
total_count = views.count()
total_count
```

```python
#qiymati bo'yicha vaqtni sortlash
views_sorted = views.sort_values(by=['hour', 'minute', 'second'])
```

```python
# min / max / mode 
min_hour = views['hour'].min()
max_hour = views['hour'].max()
mode_daytime = views['daytime'].mode()
```

```python
night_max = views[views['daytime'] == 'night']['hour'].max()
```

```python
morning_df = views[views['daytime'] == 'morning']

morning_min = morning_df['hour'].min()
morning_user = morning_df[morning_df['hour'] == morning_min].index[0]
```

```python
mode_hour = views['hour'].mode()
mode_daytime = views['daytime'].mode()
```

```python
earliest = views.nsmallest(3, 'hour')[['hour']]
latest = views.nlargest(3, 'hour')[['hour']]

earliest
latest
```
