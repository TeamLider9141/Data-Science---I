```python
import pandas as pd

df = pd.read_csv('data/auto.csv', index_col='ID')
```

```python
df.count()
```

```python
df = df.drop_duplicates(
    subset=['CarNumber', 'Make_n_Model', 'Fines'],
    keep='last')
```

```python
df.count()
```

```python
df.isna().sum()
```

```python
df = df.dropna(thresh=len(df) - 500, axis=1)

# CarNumber ustunidagi noyob qiymatlar soni
# thresh = minimum non-null deganidir. Agar ustunda 500 dan kam bo'sh qiymat bo'lsa, u ustun
# saqlanadi. Aks holda, u ustun o'chiriladi.
```

```python
df.isna().sum()
```

```python
df['Refund'] = df['Refund'].ffill()
```

```python
df.isna().sum()
```

```python
mean_fines = df['Fines'].mean()
df['Fines'] = df['Fines'].fillna(mean_fines)
```

```python
df.isna().sum()
```

```python
df['Make'] = df['Make_n_Model'].apply(lambda x: x.split()[0] if len(x.split()) > 0 else '')
df['Model'] = df['Make_n_Model'].apply(lambda x: x.split()[1] if len(x.split()) > 1 else '')
```

```python
df = df.drop(columns=['Make_n_Model'])
```

```python
df.to_json('data/auto.json', orient='records')
```
