1. Data load + index

```python
import pandas as pd

df = pd.read_json('data/auto.json')
df.set_index('CarNumber', inplace=True)
```


2. Selections (filtrlar)
Model, Make, Fines, Refund ustunlari bilan ishlaymiz.

```python
# fines > 2100
df[df['Fines'] > 2100]

```

```python
# fines > 2100 AND refund == 2
df[(df['Fines'] > 2100) & (df['Refund'] == 2)]

```

```python
# models in list
df[df['Model'].isin(['Focus', 'Corolla'])]

```

```python
# car numbers in list (index bo‘lgani uchun loc ishlatamiz)
df.loc[['Y7689C197RUS', '92928M178RUS', '7788KT197RUS',
        'H115YO163RUS', 'X758HY197RUS']]
```

3. Aggregations (make, model)

```python
#Median fines by make
df.groupby('Make')['Fines'].median()
```

```python
#Median fines by make and model
df.groupby(['Make', 'Model'])['Fines'].median()
```

```python
#Count (median’ga ishonish uchun)
df.groupby(['Make', 'Model'])['Fines'].count()
```

```python
df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
```

```python
#Std (variance tushunish uchun)
df.groupby(['Make', 'Model'])['Fines'].std()
```

4. CarNumber bo‘yicha analysis

```python
#Most violations (count)
violations = df.groupby(df.index)['Fines'].count().sort_values(ascending=False)
```

```python
#Top-1 car
top_car = violations.index[0]
df.loc[top_car]
```

```python
#Most paid (sum)
paid = df.groupby(df.index)['Fines'].sum().sort_values(ascending=False)

top_paid_car = paid.index[0]
df.loc[top_paid_car]
```
