1. Imports

```python
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:.2f}'.format
```

2. JSON ni o‘qish

```python
df = pd.read_json('data/auto.json')
df.head()
```

3. Sample yaratish (200 ta)

```python
sample = df.sample(n=200, random_state=21, replace=True)

#sample
```

4. Concatenate

```python
concat_rows = pd.concat([df, sample], ignore_index=True)
```

5. Year column qo‘shish

```python
np.random.seed(21)
years = np.random.randint(1980, 2020, size=len(concat_rows))

concat_rows['Year'] = years

fines = concat_rows.copy()
```

6. Owners dataframe yaratish
=> surname.json o‘qish

```python
surnames = pd.read_json('data/surname.json')
print(surnames.columns)
print(surnames.head())
```

```python
surnames.columns = ['NAME','COUNT','RANK']
surnames_series = surnames['NAME'].str.replace(r'[^A-Za-z]', '', regex=True)
```

```python
unique_cars = fines['CarNumber'].unique()
```

```python
owners_surnames = surnames_series.sample(
    n=len(unique_cars), 
    random_state=21, 
    replace=True
).values
```

```python
owners = pd.DataFrame({
    'CarNumber': unique_cars,
    'SURNAME': owners_surnames
})
```

7. fines ga 5 ta yangi observation qo‘shish

```python
new_fines = pd.DataFrame({
    'CarNumber': ['X111XX', 'X222XX', 'X333XX', 'X444XX', 'X555XX'],
    'Make': ['BMW', 'Audi', 'Kia', 'Toyota', 'Honda'],
    'Model': ['X5', 'A6', 'Rio', 'Camry', 'Civic'],
    'Refund': [0, 1, 0, 1, 0],
    'Fines': [100.0, 200.0, 150.0, 300.0, 250.0],
    'Year': np.random.randint(1980, 2020, 5)
})

fines = pd.concat([fines, new_fines], ignore_index=True)
```

8. owners ni modify qilish

```python
owners = owners.iloc[:-20]

new_owners = pd.DataFrame({
    'CarNumber': ['Y111YY', 'Y222YY', 'Y333YY'],
    'SURNAME': ['Smith', 'Johnson', 'Williams']
})

owners = pd.concat([owners, new_owners], ignore_index=True)
```

9. Joinlar (4 xil)

```python
#1. Intersection
inner_join = pd.merge(fines, owners, on='CarNumber', how='inner')
```

```python
#2. Union
outer_join = pd.merge(fines, owners, on='CarNumber', how='outer')
```

```python
#3. Fines only
left_join = pd.merge(fines, owners, on='CarNumber', how='left')
```

```python
#4. Owners only
right_join = pd.merge(fines, owners, on='CarNumber', how='right')
```

10. Pivot table

```python
pivot = fines.pivot_table(
    values='Fines',
    index='CarNumber',
    columns='Year',
    aggfunc='sum',
    fill_value=0
)
```

11. CSV ga saqlash

```python
fines.to_csv('fines.csv', index=False)
owners.to_csv('owners.csv', index=False)
```
